```python
#IMPORT SECTION 
import cv2 
import numpy as np 
import matplotlib as mpl 
from matplotlib import pyplot as plt 
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Load the image
image_path = "/Users/my475/OneDrive/Desktop/image 1.jpg"

# Check if the image exists at the specified path
if os.path.exists(image_path):
    building = cv2.imread(image_path)
    
    # Check if the image was loaded successfully
    if building is not None:
        # Convert the color from BGR to RGB
        building_rgb = cv2.cvtColor(building, cv2.COLOR_BGR2RGB)
        
        # Display the image using matplotlib
        plt.imshow(building_rgb)
        plt.axis('off')  # To hide the axes
        plt.show()
    else:
        print("Error: The image could not be loaded. Please check the file.")
else:
    print("Error: The file does not exist. Please check the file path.")
```

```python
#MEAN FILTER 3x3
bg3 = cv2.GaussianBlur(building,(3,3),0) 
```

```python
#MEAN FILTER 5x5 
bg5 = cv2.GaussianBlur(building,(5,5),0) 
```

```python
#MEAN FILTER 7x7 
bg7 = cv2.GaussianBlur(building,(7,7),0) 
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=[10,10]) 
ax[0].imshow(building) 
ax[0].set_title("Original Image") 
ax[1].imshow(bg3) 
ax[1].set_title("3X3 Blur") 
ax[2].imshow(bg5) 
ax[2].set_title("5X5 Blur") 
ax[3].imshow(bg7) 
ax[3].set_title("7X7 Blur") 
fig.tight_layout() 
plt.show() 
```

```python
import cv2
import matplotlib as mpl
from matplotlib import pyplot as plt
img = cv2.imread("/Users/my475/OneDrive/Desktop/image 1.jpg")
im1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
im1_yuv = cv2.cvtColor(im1,cv2.COLOR_BGR2YUV)
im1_yuv[:,:,0] = cv2.equalizeHist(im1_yuv[:,:,0])
hist_eq = cv2.cvtColor(im1_yuv, cv2.COLOR_YUV2BGR)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[25,25])
ax[0].imshow(im1)
ax[0].set_title('ORIGINAL IMAGE')
ax[1].imshow(hist_eq)
ax[1].set_title('EQUALIZING \nUSING YUV MODEL')
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("/Users/my475/OneDrive/Desktop/image 1.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(image_rgb)
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)
eq_image = cv2.merge([r_eq, g_eq, b_eq])
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(eq_image)
ax[1].set_title('Equalized Image')
ax[1].axis('off')
ax[2].imshow(eq_image)
ax[2].set_title('EQUALIZING \nEACH CHANNEL')
ax[2].axis('off')
plt.tight_layout()
plt.show()
```
